```python
from scipy.stats import mannwhitneyu
```

```python
cond1 = kibble_lr_named[kibble_lr_named['diet_Dry_kibble']=='yes'].Current_Natural_Log_Ratio.values
cond2 = kibble_lr_named[kibble_lr_named['diet_Dry_kibble']=='no'].Current_Natural_Log_Ratio.values
stat, pval = mannwhitneyu(cond1, cond2, method="auto", alternative='two-sided')
print("Comparison of LR values eating kibble")
print("p-value: ", pval)
```

```python
cond1 = df[df['diet_form']=='kibble only'].Current_Natural_Log_Ratio.values
cond2 = df[df['diet_form']=='canned wet food only'].Current_Natural_Log_Ratio.values
cond3 = df[df['diet_form']=='both'].Current_Natural_Log_Ratio.values
```

```python
stat1, pval1 = mannwhitneyu(cond1, cond2, method="auto", alternative='two-sided')
print("Comparison of LR values eating kibble only to wet food only")
print("p-value: ", pval1)
print("stat: ", stat1)
```

```python
stat2, pval2 = mannwhitneyu(cond1, cond3, method="auto", alternative='two-sided')
print("Comparison of LR values eating kibble only to both")
print("p-value: ", pval2)
print("stat: ", stat2)
```

```python
stat3, pval3 = mannwhitneyu(cond2, cond3, method="auto", alternative='two-sided')
print("Comparison of LR values eating wet food only to both")
print("p-value: ", pval3)
print("stat: ", stat3)
```

```python
from statsmodels.sandbox.stats.multicomp import multipletests
```

```python
pvals = [pval1, pval2, pval3]
```

```python
p_adjusted = multipletests(pvals, method='bonferroni')
print(p_adjusted)
#[4.08650682e-06, 0.0760049083, 4.92930285e-06]
```

```python
p_adjusted = multipletests(pvals, method='holm')
print(p_adjusted)
#[4.08650682e-06, 0.02.53349694, 4.08650682e-06]
```

```python
p_adjusted = multipletests(pvals, method='fdr_bh')
print(p_adjusted)
#[2.46465143e-06, 0.0253349694, 2.46465143e-06]
```
